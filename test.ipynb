# preprocess the data

```python
import pandas as pd
from datetime import date
import calendar
import xlrd
import datetime
```

```python
df=pd.read_excel('atm_new.xlsx',sheet_name='Sheet 1')
```

```python
df.head()
```

```python
df.shape
```

```python
df['TXN_DATE'] =  pd.to_datetime(df['TXN_DATE'])
```

```python
df['trxn_year'] = pd.to_datetime(df['TXN_DATE']).dt.year
df['trxn_month'] = pd.to_datetime(df['TXN_DATE']).dt.month
df['trxn_week'] = pd.to_datetime(df['TXN_DATE']).dt.week
#df['trxn_week_pre'] = pd.to_datetime(df['TXN_DATE']).dt.week-1
df['trxn_date'] = pd.to_datetime(df['TXN_DATE']).dt.weekday
df['date'] = pd.to_datetime(df['TXN_DATE']).dt.day_name()
```

# previous week

```python
df_2=df.groupby(['TERMINAL_ID','trxn_year','trxn_week'])['WITHDRAWAL'].sum()
df_2 =df_2.reset_index()
df_2.rename(columns= {'WITHDRAWAL':'week_total'}, inplace =True)
```

```python
df_2
```

```python
df= df.merge(df_2, on = ['TERMINAL_ID', 'trxn_year', 'trxn_week'])
```

```python
df
```

```python
df_3=df.groupby(['TERMINAL_ID','trxn_year','trxn_week'])['WITHDRAWAL'].mean()
df_3 =df_3.reset_index()
df_3.rename(columns= {'WITHDRAWAL':'week_mean'}, inplace =True)
```

```python
df_3['prv_week']= df_3.groupby(['TERMINAL_ID', 'trxn_year'])['week_mean'].shift(1)
```

```python
df_3['Pr_mnt_crrs_week'] =  df_3.groupby(['TERMINAL_ID', 'trxn_year'])['week_mean'].shift(4)
```

```python
df_3
```

```python
df = df.merge(df_3, on=['TERMINAL_ID','trxn_year', 'trxn_week'])
```

```python
df
```

# holiday and holiday sequence

```python
hol=pd.read_excel('holiday.xlsx')
hol
```

```python
import datetime
hol['prv _date']=hol['TXN_DATE']- datetime.timedelta(days=1)
```

```python
df['HOLIDAY']='W'
```

```python
df = df.merge(hol,on='TXN_DATE',how="left")
```

```python
df['HOLIDAY_y'] = df['HOLIDAY_y'].fillna(df['HOLIDAY_x'])
df.drop(['HOLIDAY_x'],inplace=True,axis=1)
df.rename(columns={'HOLIDAY_y':'HOLIDAY'},inplace=True)
```

```python
df['prv_day']= df.groupby(['TERMINAL_ID', 'trxn_year'])['HOLIDAY'].shift(1)
df['nxt_day']= df.groupby(['TERMINAL_ID', 'trxn_year'])['HOLIDAY'].shift(-1)
```

```python
df.drop(['prv _date'],inplace=True,axis=1)
```

```python
df[['prv_day', 'nxt_day']]=df[['prv_day', 'nxt_day']].fillna('W')
```

```python
df['hol_sequence']=df['prv_day']+df['HOLIDAY']+df['nxt_day']
```

```python
df
```

```python
df=df[['DISTRICTNAME','BRANCHNAME','TXN_DATE','TERMINAL_ID','trxn_year','trxn_month','trxn_week','trxn_date', 
           'date','WITHDRAWAL','week_mean','Pr_mnt_crrs_week','prv_week','HOLIDAY','hol_sequence']]
```

```python
df['week_mean'].isnull().values.any()
```

# prv month

```python
df_4=df.groupby(['TERMINAL_ID','trxn_year','trxn_month'])['WITHDRAWAL'].mean()
df_4 =df_4.reset_index()
df_4.rename(columns= {'WITHDRAWAL':'month_mean'}, inplace =True)
```

```python
df_4['prv_month']= df_4.groupby(['TERMINAL_ID', 'trxn_year'])['month_mean'].shift(1)
```

```python
df = df.merge(df_4, on=['TERMINAL_ID','trxn_year', 'trxn_month'])
```

```python
df
```

```python
df['month_mean'].isnull().values.any()
```

```python
df[df.isna().any(axis=1)]
```

```python
df['prv_week'] = df.prv_week.fillna(df.groupby(['trxn_year','TERMINAL_ID','BRANCHNAME'])['prv_week'].transform('mean')) 
df['prv_month'] = df.prv_month.fillna(df.groupby(['trxn_year','TERMINAL_ID','BRANCHNAME'])['prv_month'].transform('mean'))
df['Pr_mnt_crrs_week'] = df.Pr_mnt_crrs_week.fillna(df.groupby(['trxn_year','TERMINAL_ID','BRANCHNAME'])['Pr_mnt_crrs_week'].transform('mean'))
```

```python
df[df['prv_week'].isna()].shape
```

```python
df=df.dropna()
```

```python
data = df.set_index('TXN_DATE')
```

```python
data
```

```python
district_ref=pd.read_excel('reference.xlsx',sheet_name='district')
branch_ref=pd.read_excel('reference.xlsx',sheet_name='branch')
terminal_ref=pd.read_excel('reference.xlsx',sheet_name='terminal')
holiday_ref=pd.read_excel('reference.xlsx',sheet_name='holiday')
holseq_ref=pd.read_excel('reference.xlsx',sheet_name='hol_seq')
```

```python
data = data.merge(district_ref,on='DISTRICTNAME',how="left")
data = data.merge(branch_ref,on='BRANCHNAME',how="left")
data = data.merge(terminal_ref,on='TERMINAL_ID',how="left")
data = data.merge(holiday_ref,on='HOLIDAY',how="left")
data = data.merge(holseq_ref,on='hol_sequence',how="left")
```

```python
data
```

```python
input=data[['prv_week','prv_month','Pr_mnt_crrs_week','District','Branch','Terminal','trxn_month',
            'trxn_week','trxn_date','holiday','hol_seq' ]]
```

```python
input
```

# load model

```python
from joblib import Parallel, delayed
import joblib
import pandas as pd
```

# for linear regression

```python
#Load the model from the file
linear_reg = joblib.load('linear_reg2.pkl')
  
```

```python
lin_pred=linear_reg.predict(input)
```

```python
linpred = pd.DataFrame(lin_pred[:]) 
linpred.rename(columns = {0: 'lin_predicted'}, inplace=True) # renaming the column
linpred = linpred.round(decimals=0) # rounding the decimal values
d = pd.DataFrame(input) # calling last 50 values of original amt wothdrawn
linpred.index = d.index # mapping the index of both dataframe
linok = pd.concat([linpred, d], axis=1) 
linok
```

```python
Result=linok
```

```python
Result = Result.merge(district_ref,on='District',how="left")
Result = Result.merge(branch_ref,on='Branch',how="left")
Result = Result.merge(terminal_ref,on='Terminal',how="left")
```

```python
Result
```

```python
Result.drop(['prv_week','prv_month','Pr_mnt_crrs_week','District','Branch','Terminal',
             'trxn_month','trxn_week','trxn_date','HOLIDAY','hol_sequence'],inplace=True,axis=1)
```

```python
Result
```

```python
Result.to_csv('linear_reg_pridivtion.csv')
```
